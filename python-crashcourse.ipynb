You can also use comments to tell the python interpreter to ignore the piece of code. It is not a recommended practice though.

## Repetition (Looping)

Repetitions are an important control flow structure in any programming language, it allows you to execute certain code blocks repeatedly. Loopings in python can be achieved by using `while` and `for` statements.

### while statement

While is a type of control flow that is used when you want to execute some code blocks until a desired condition is met. Let's print something three times using while loop:

```python
counter = 0
while counter<3:
  print("Something")
  counter+=1           #a = a+1 is equivalent to a+=1
```

````python
import random                                     #ignore this line for now
n = random.randrange(1,10)                        #ignore this line too
guess = int(input("Enter any number: "))
while n!= guess:
    if guess < n:
        print("Too low")
        guess = int(input("Enter number again: "))
    elif guess > n:
        print("Too high!")
        guess = int(input("Enter number again: "))
    else:
      break
print("you guessed it right!!")
````

# Functions
Functions are a block of code that performs some 'function'. We use them so that we may reuse the code. Let's say we want to perform a series of mathematical operations on variables `a` and `b`.

- add `a` and `b` and assign to a new variable `sum`
- divide the `sum` by 2.

## Parts of a function.
Functions have :
- Function Name: For any function, it should have it's own unique name.
- Parameters or Arguments: These are the inputs to the function. for example `a` and `b` are the parameters. Generally, some kind of operations are performed using these parameters.

- body: Of course, the body is where we write down what exactly the code does.

Optionals:
- return statement: what the function returns. In our example, it returns a number after some arithmatic operations.
- Docstring: The programmer can use docstring to explain what the code does. ( Comments but inside a function!)


````python
def operation(a: int, b: int)->float:
  """
  This function performs some arithmatic operations.

  parameters:
    a: first number
    b: second number
  returns:
    a number
  """
  sum = a+b
  sum = sum/2.0
  return sum
````

Extra tip:

Although the python is a dynamically typed language, it is a good practice to use typing while constructing functions. See the function definition in the above example.

```python
def operation(a: int, b: int)->float:
```

In this line, we specify the data types of the parameters and also the datatype of what the function is returning (It's only for providing hints and does nothing) 

````python
operation(1,6)
````

````python
import cv2, requests, numpy as np
from google.colab.patches import cv2_imshow

req = requests.get('https://i2-prod.mirror.co.uk/incoming/article25609246.ece/ALTERNATES/s1200d/0_PUSS-IN-BOOTS.jpg')
arr = np.asarray(bytearray(req.content),)
img = cv2.imdecode(arr, -1)
cv2_imshow(img)
````

````python
bw_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
````
